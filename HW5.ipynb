```r
install.packages("tidyverse")
install.packages("googledrive")
install.packages("ggplot2")
install.packages("dplyr")
library(tidyverse)
library(googledrive)
library(ggplot2)
drive_auth()
drive_find(pattern = "house_prices.rda")
drive_download ("house_prices.rda", path = "house_prices.rda", overwrite = TRUE)
load("house_prices.rda")
ls()
#I was having issues loading google drive so I used chat gpt for guidance
```

```r
colnames(house_prices)

ggplot(house_prices, aes(x = date, y = house_price_index)) +
  geom_line() +
  facet_wrap(~state, scales = "free_x") + #I used chatgpt to figure out how to have the ticks on each states x-axis
  scale_x_continuous(
    breaks = c(1980, 2000, 2020),
    labels = c("'80", "'00", "'20")
  ) +
  labs(
    title = "House Price Index Trend Over Time By State",
    x = "Year", y = "House Price Index"
  )

house_reshaped <- house_prices %>%
  gather(key = "measure", value = "value", -state, -date)

```

```r
ggplot(house_reshaped, aes(x = date, y = value, color = measure)) +
  geom_line() +
  facet_wrap(~state, scales = "free_x" ) +
  scale_x_continuous(
    breaks = c(1980, 2000, 2020),
    labels = c("'80", "'00", "'20")
  ) +
  labs(
    title = "House Price Index & Unemployment Percent Over Time By State",
    x = "Year", y = "Value", color = "Measurement"
  )
```
