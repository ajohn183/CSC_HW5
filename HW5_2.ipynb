```r
upload <- file.choose()
load("/content/house_prices.rda")

library(dplyr)
library(tidyr)
library(ggplot2)
library(lubridate)

house_prices %>%
  gather(key = measure, value = value, -c(house_price_index, date, state)) -> house_reshaped

ggplot(house_reshaped, aes(x = date, y = value, color = measure)) +
  geom_line() +
  facet_wrap(~state) +
  scale_x_continuous(
    breaks = ymd(c("1980-01-01", "2000-01-01", "2020-01-01")),
    labels = c("80", "00", "20")
  ) +
  labs(
    title = "Unemployment & House Price Index Change Over Time By State",
    x = "Year", y = "Percent", color = "Measurement"
  )
#differences: X-axis now includes all three ('80, '00, and '20)
```
